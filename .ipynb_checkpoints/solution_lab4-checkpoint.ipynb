## Roll a dice 5 * 10<sup>4</sup> times and plot the frequency distribution

````python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import random
````

## Using ```random.randint``` 

````python
list1 = [random.randint(1, 6) for i in range(5*10**4)]
freq1 = [0] * 7
for i in range(1, 7):
    print('{} - {}'.format(i, list1.count(i)))
    freq1[i] = float(list1.count(i))/50000
del freq1[0]
````

### Plotting using ```matplotlib.pyplot```

````python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].hist(list1)
ax[1].bar([1, 2, 3, 4, 5, 6], freq1)
````

### Plotting Using ```seaborn```

````python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
sns.countplot(list1, ax = ax[0])
sns.barplot(x = [1, 2, 3, 4, 5, 6], y = freq1, ax = ax[1])
ax[0].set(xlabel = "Event", ylabel = "Frequency", title = "Frequency Distribution Graph")
ax[1].set(xlabel = "Event", ylabel = "Probability", title = "Probability Distribution Graph")
````
