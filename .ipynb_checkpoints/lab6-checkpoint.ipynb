## Binomial Distribution Graph

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# hcount = []
# tcount = []
hfreq  = [0] * 6
tfreq  = [0] * 6
```

```python
for i in range(50000):
    a = [random.choice(['H', 'T'])for i in range(5)]
#     hcount.append(a.count('H'))
#     tcount.append(a.count('T'))
    hfreq[a.count('H')] += 1
    tfreq[a.count('T')] += 1
```

```python
hdata = np.array(hfreq).astype(float) / 50000
tdata = np.array(tfreq).astype(float) / 50000
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
sns.barplot(x = [0, 1, 2, 3, 4, 5], y = hdata, ax = ax[0])
ax[0].set(ylabel = "Probability", xlabel = "Event")
ax[0].set_title("Heads Binomial Distribution")
sns.barplot(x = [0, 1, 2, 3, 4, 5], y = tdata, ax = ax[1])
ax[1].set_title("Tails Binomial Distribution")
ax[1].set(ylabel = "Probability", xlabel = "Event")

# plt.yticks(np.array(hfreq).astype(float) / 50000)
```

## Assignment

```python
def ncr(n, r):
    a, b = max(n - r, r), min(n - r, r)
    numerator = 1
    for i in range(n, a, -1): numerator *= i
    denominator = 1
    for i in range(1, b + 1):denominator *= i
    return numerator / denominator
```

```python
p, q, n = 0.68, 0.32, 5
```

```python
def plot(n, p, q):
    data = [ncr(n, r) * (p**r) * (q**(n-r)) for r in range(0, n + 1)]
    
    fig, ax = plt.subplots(figsize = (12, 7))
    sns.barplot(x = [i for i in range(n + 1)], y = data)
    ax.set(xlabel = "Event", ylabel = "Probability")
    plt.title("Binomial Probability Distrbution Graph")
```

```python
plot(n, q, p)
```
