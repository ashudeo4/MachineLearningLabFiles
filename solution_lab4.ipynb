## Roll a dice 5 * 10<sup>4</sup> times and plot the frequency distribution

````python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import random
````

### Using ```random.randint``` 

````python
list1 = [random.randint(1, 6) for i in range(5*10**4)]
freq = [0] * 7
for i in range(1, 7):
    print('{} - {}'.format(i, list1.count(i)))
    freq[i] = float(list1.count(i))/50000
del freq[0]
````

### Plotting using ```matplotlib.pyplot```

````python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].hist(list1)
ax[1].bar([1, 2, 3, 4, 5, 6], freq)
````
