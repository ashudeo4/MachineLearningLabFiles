## Question : Roll a dice 5 * 10<sup>4</sup> times and plot the frequency distribution

````python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import random
````

## 1. Using ```random.randint``` 

````python
averages1 = []
for i in range(50000):
    list1 = [random.randint(1, 6) for i in range(5)]
    averages1.append(np.mean(list1))
freq1 = {}
for i in averages1:
    if i not in freq1.keys(): freq1[i] = 1
    else : freq1[i] += 1
````

   ### 1.1. Plotting using ```matplotlib.pyplot```

````python
fig, ax = plt.subplots(figsize = (15, 7))
plt.bar(x = freq1.keys(), height = freq1.values())
plt.xticks(freq1.keys())
plt.xlabel("Averages")
plt.ylabel("Frequency")
plt.title("Probability Distribution Graph")
plt.show()
````

### 1.2. Plotting Using ```seaborn```

````python
fig, ax = plt.subplots(figsize = (15, 7))
sns.barplot(x = freq1.keys(), y = freq1.values())
ax.set(xlabel = "Average", ylabel = "Frequency", title = "Probability Distribution Graph")
````
