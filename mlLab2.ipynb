### Mean = 72, Std = 5

```python
import random

data = []
mean = 72
std = 5
for i in range(50000):
    data += [mean+random.uniform(-4, 4)*std]

# Z-score    
```

```python
print(data)
```

```python
import numpy as np
data = np.array(data)
```

```python
np.mean(data)
```

```python
np.std(data)
```

```python
# bubble sort
arr = [2,6,1,3,9,1]
for i in range(len(arr)):
    for j in range(len(arr)-1):
        if arr[j+1] < arr[j]:
            arr[j+1], arr[j] = arr[j], arr[j+1]
```
